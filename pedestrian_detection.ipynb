# Detecting Pedestrian

```python
import cv2
```

```python
import numpy as np
```

```python
# create body classifier
body_classifier = cv2.CascadeClassifier("haarcascade_fullbody.xml")
```

```python
# initiate video capture for video file
cap = cv2.VideoCapture("walking.avi")
# cap = cv2.VideoCapture("cctv.mp4")
```

```python
# loop once video is successfully loaded
while cap.isOpened():
    
    # read first frame
    ret, frame = cap.read()
    frame = cv2.resize(frame, None, fx=1, fy=1, interpolation = cv2.INTER_LINEAR)
    grey = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    # pass frame into body classifier
    bodies = body_classifier.detectMultiScale(grey, 1.2, 3)
    
    # Extract bounding boxes for any bodies identified
    for (x, y, w, h) in bodies:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 255), 2)
        cv2.imshow("Pedestrians", frame)
    
    # break loop if "Enter" key is pressed
    if cv2.waitKey(1) == 13:
        break
        
cap.release()
cv2.destroyAllWindows()
```
